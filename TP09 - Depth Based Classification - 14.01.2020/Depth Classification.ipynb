```python
import numpy as np 
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
mean_1 = [0,0]
cov_1 = [[1,1],[1,4]]
MVN1= np.random.multivariate_normal(mean_1,cov_1,25).T

mean_2 = [2,2]
cov_2 = [[1,1],[1,4]]
MVN2= np.random.multivariate_normal(mean_2,cov_2,25).T

mean_3 = [2,2]
cov_3 = [[4,4],[4,16]]
MVN3= np.random.multivariate_normal(mean_3,cov_3,25).T
```

```python
x1,y1 = MVN1
x2,y2 = MVN2
x3,y3 = MVN3
```

```python
plt.plot(x1,y1, 'x')
plt.axis('equal')
plt.show()

plt.plot(x2,y2, 'x')
plt.axis('equal')
plt.title("MVN2")
plt.show()

plt.plot(x3,y3, 'x')
plt.axis('equal')
plt.show()
```

```python
def depthMah (x, X):
    
    """
    Profondeur de MAHALANOBIS
    """
    #sample mean
    mu = np.mean(X, axis =1)
    
    #sample covariance
    cov = np.cov(X)
    
    ## Mahalanobis distance
    d_mah = np.dot(np.dot((x.T-mu),inv(cov)).T,(x.T-mu))
    
    ## Mahalanobis depth
    depth_mah = 1/(1+ d_mah) 
    
    return depth_mah
    
    
```

```python
o = MVN1.T[0]
```

```python
depthMah(o,MVN1)
```

```python
list_MVN1 = []
for o in MVN1.T:  #o un point du dataset 
    list_MVN1.append(depthMah(o,MVN1))
    
list_MVN2 = []
for o in MVN2.T:  #o un point du dataset 
    list_MVN2.append(depthMah(o,MVN2))
    
list_MVN3 = []
for o in MVN3.T:  #o un point du dataset 
    list_MVN3.append(depthMah(o,MVN3))
```

```python
plt.plot(x1,y1, 'x')
for i, txt in enumerate(list_MVN1):
    plt.annotate(txt,MVN1.T[i])
    
```
